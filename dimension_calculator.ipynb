## Mobility Scooter Dimension Calculator
The following calculator takes anthropometric measurements and joint angles as inputs and outputs the optimal dimensions A, B, C and D (all of which use the end of the foot plate as reference).

```python
from tools_new import joint_positions_calculator, calculator_plot, calculate_distances
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
# Input back inclination angle
back_inclination = 90
ankle_angle = 90                # Ankle Flexion
knee_angle = 93                 # Knee Flexion
shoulder_angle = 25             # Shoulder Flexion

# Input anthropometric measurements
foot_length = 240 - (52.9)
lower_leg_length = 580 - (75.25)
thigh_length = 500 - (160.25)
torso_height = 700 - (160.25)
upper_arm_length = 450
forearm_length = 400 - (108.55)

angles_in = [ankle_angle, knee_angle, back_inclination, shoulder_angle]
```

```python
# Calculate and plot joints
joints = joint_positions_calculator(foot_length, lower_leg_length, thigh_length, torso_height, upper_arm_length, forearm_length, angles_in)
calculator_plot(joints, dists = True)

# Calculate the required distances
distance_A, distance_B, distance_C, distance_D = calculate_distances(joints)

# Output the distances
print(f"A (Horizontal distance): {distance_A: .2f} mm")
print(f"B (Vertical distance): {distance_B: .2f} mm")
print(f"C (Vertical distance): {distance_C: .2f} mm")
print(f"D (Horizontal distance): {distance_D: .2f} mm")
```
