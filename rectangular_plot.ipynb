```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load the image (replace 'your_image.png' with the actual file path)
image_path = "assets/3Scooter Simplified CAD v1 Side View.png"  # Ensure this file exists
img = mpimg.imread(image_path)

# Define rectangle coordinates
rectangles = [
    {"x_min": -790, "x_max": -500, "y_min": 284, "y_max": 542},
    {"x_min": -295, "x_max": 0, "y_min": 431, "y_max": 705},
]

# Define datapoints
datapoints = [(0, 0), (-600, 360), (-66, 639)]

# Define figure and axis
fig, ax = plt.subplots(figsize=(8, 6), dpi=300)

img_height, img_width = img.shape[:2]
img_aspect = (img_width / img_height) * 1.2

# Pick a data width that fits your coordinate system (e.g., 1000)
x_min, x_max = -880, 220
data_width = x_max - x_min
data_height = data_width / img_aspect
y_min, y_max = -190, data_height-190

# Plot image
ax.imshow(img, extent=[x_min, x_max, y_min, y_max], aspect='equal')

# Plot filled rectangles with 50% opacity and no borders, with colors and labels
colors = ['orange', 'green']
labels = ['Seat Adjustability Area', 'Handlebar Adjustability Area']

for rect, color, label in zip(rectangles, colors, labels):
    width = rect["x_max"] - rect["x_min"]
    height = rect["y_max"] - rect["y_min"]
    ax.add_patch(plt.Rectangle((rect["x_min"], rect["y_min"]), width, height, 
                               facecolor=color, edgecolor='none', alpha=0.3, label=label))

# Plot datapoints individually with different colors
joint_x, joint_y = zip(*datapoints)
colors = ['red', 'orange', 'lightgreen']

for x, y, c in zip(joint_x, joint_y, colors):
    ax.scatter(x, y, color=c, edgecolors='black', linewidths=0.5, marker='o', s=70)

# Horizontal A
ax.annotate('', 
    xy=(joint_x[2], joint_y[0] + 750),  # Start of the line (xy)
    xytext=(joint_x[0], joint_y[0] + 750),  # End of the line (xytext)
    arrowprops=dict(facecolor='black', arrowstyle='<-'))  # Arrow at the start
ax.annotate('', 
    xy=(joint_x[2], joint_y[0] + 750),  # Start of the line (xy)
    xytext=(joint_x[0], joint_y[0] + 750),  # End of the line (xytext)
    arrowprops=dict(facecolor='black', arrowstyle='->'))  # Arrow at the end
ax.vlines(joint_x[2], joint_y[2], joint_y[0]+750 , linestyle='--', colors='black', linewidth=0.5)
ax.vlines(joint_x[0], joint_y[0], joint_y[0]+750 , linestyle='--', colors='black', linewidth=0.5)
ax.text((joint_x[2] + joint_x[0]) / 2, joint_y[0] + 770, "Distance A", ha='center', va='bottom', fontsize=10)

# Vertical B
ax.annotate('', 
    xy=(joint_x[0] + 300, joint_y[0]),  # Start of the line (xy)
    xytext=(joint_x[0] + 300, joint_y[2]),  # End of the line (xytext)
    arrowprops=dict(facecolor='black', arrowstyle='<-'))  # Arrow at the start
ax.annotate('', 
    xy=(joint_x[0]+ 300 , joint_y[0]),  # Start of the line (xy)
    xytext=(joint_x[0] + 300, joint_y[2]),  # End of the line (xytext)
    arrowprops=dict(facecolor='black', arrowstyle='->'))  # Arrow at the end
ax.hlines(joint_y[0], joint_x[0], joint_x[0]+300 , linestyle='--', colors='black', linewidth=0.5)
ax.hlines(joint_y[2], joint_x[2], joint_x[0]+300 , linestyle='--', colors='black', linewidth=0.5)
ax.text(joint_x[0] + 325, (joint_y[2] + joint_y[0]) / 2, "Distance B", ha='left', va='center', fontsize=10, rotation=90)

# Vertical C
ax.annotate('', 
    xy=(joint_x[0] - 900, joint_y[0]),  # Start of the line (xy)
    xytext=(joint_x[0] - 900, joint_y[1]),  # End of the line (xytext)
    arrowprops=dict(facecolor='black', arrowstyle='<-'))  # Arrow at the start
ax.annotate('', 
    xy=(joint_x[0] - 900, joint_y[0]),  # Start of the line (xy)
    xytext=(joint_x[0] - 900, joint_y[1]),  # End of the line (xytext)
    arrowprops=dict(facecolor='black', arrowstyle='->'))  # Arrow at the end
ax.hlines(joint_y[0], joint_x[0], joint_x[0]-900 , linestyle='--', colors='black', linewidth=0.5)
ax.hlines(joint_y[1], joint_x[1], joint_x[0]-900 , linestyle='--', colors='black', linewidth=0.5)
ax.text(joint_x[0] - 945, (joint_y[1] + joint_y[0]) / 2, "Distance C", ha='left', va='center', fontsize=10, rotation=90)

# Horizontal D
ax.annotate('', 
    xy=(joint_x[1], joint_y[0] - 145),  # Start of the line (xy)
    xytext=(joint_x[0], joint_y[0] - 145),  # End of the line (xytext)
    arrowprops=dict(facecolor='black', arrowstyle='<-'))  # Arrow at the start
ax.annotate('', 
    xy=(joint_x[1], joint_y[0] - 145),  # Start of the line (xy)
    xytext=(joint_x[0], joint_y[0] - 145),  # End of the line (xytext)
    arrowprops=dict(facecolor='black', arrowstyle='->'))  # Arrow at the end
ax.vlines(joint_x[1], joint_y[1], joint_y[0]-145 , linestyle='--', colors='black', linewidth=0.5)
ax.vlines(joint_x[0], joint_y[0], joint_y[0]-145 , linestyle='--', colors='black', linewidth=0.5)
ax.text((joint_x[1] + joint_x[0]) / 2, joint_y[0] - 200, "Distance D", ha='center', va='bottom', fontsize=10)

# Set axis limits
ax.set_xlim(-1100, 550)
ax.set_ylim(-230, 850)

# Make x and y axis use the same scale
ax.set_aspect('equal')

ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_title("Handlebar and Seat Areas of Adjustability")

# Show grid and legend
ax.grid(True, linestyle='--', alpha=0.2)
ax.legend()

# Show plot
plt.show()
```
