```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


df = pd.read_csv('raw_data/SlovakiaDataset.csv')

df.columns
```

```python
data_new = pd.DataFrame()

data_new['Gender'] = df['Gender']
data_new['FootLength'] = df['25.Foot lenght (cm)']
data_new['LowerLegLength'] = df['16.Knee height, sitting (cm)']
data_new['ThighLength'] = df['19.Buttock-knee length (cm)']
data_new['TorsoHeight'] = df['14.Elbow height, sitting (cm)'] + df['15.Shoulder-elbow length (cm)']
data_new['UpperArmLength'] = df['15.Shoulder-elbow length (cm)']
data_new['ForearmLength'] = df['18.Forearm-fingertip length  (cm)'] - (2/3)*(df['23.Palm length (cm)'])
```

```python
data_new.dropna(inplace=True)
data = data_new.drop(columns = 'Gender')

# Scale data before applying PCA
scaling=StandardScaler()

# Use fit and transform method 
scaling.fit(data)
Scaled_data=scaling.transform(data)

# Set the n_components=2
principal=PCA(n_components=2)
principal.fit(Scaled_data)
x=principal.transform(Scaled_data)

# Check the dimensions of data after PCA
print(x.shape)
```

```python
print(principal.explained_variance_ratio_)  # Proportion of variance explained
print(sum(principal.explained_variance_ratio_))  # Total variance explained
```

```python
data_new['Gender'] = data_new['Gender'].map({'M': 'tab:blue', 'F': 'tab:pink'})  # Convert to numeric

plt.figure(figsize=(10,10))
plt.scatter(x[:,0],x[:,1],c=data_new['Gender'],cmap='plasma')
plt.xlabel('pc1')
plt.ylabel('pc2')
```

```python
# Get the PCA loadings (eigenvectors)
loadings = pd.DataFrame(principal.components_, columns=data.columns, index=[f'PC{i+1}' for i in range(principal.n_components_)])
loadings
```

```python
import seaborn as sns

plt.figure(figsize=(12, 10))
sns.heatmap(loadings, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("PCA Component Loadings")
plt.show()
```

```python
import pandas as pd

# Create a DataFrame with PC1 and PC2
df_pca = pd.DataFrame(x[:, :2], columns=['PC1', 'PC2'], index=data.index)  # Keep original indices

# Add Gender back to the PCA DataFrame for easy lookup
df_pca['Gender'] = data_new['Gender']

# Now df_pca contains PC1, PC2, and Gender!
df_pca.head()
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np

# Function to get original data when clicking on points
def onpick(event):
    ind = event.ind  # Get index of selected points
    selected_points = df_pca.iloc[ind]  # Retrieve selected rows
    print("\nSelected Data Points (Original Indices):")
    print(selected_points)  # Print original data

# Plot the PCA scatter plot with interactive picking
fig, ax = plt.subplots(figsize=(8, 6))
sc = ax.scatter(df_pca['PC1'], df_pca['PC2'], alpha=0.5, c=data_new['Gender'], picker=True)

ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_title("Interactive PCA Scatter Plot")

fig.canvas.mpl_connect('pick_event', onpick)  # Connect the picking event

plt.show()
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

# Example Data
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)

# Function to print selected points
def onpick(event):
    ind = event.ind  # Get selected indices
    print(f"\nSelected Point(s) Index: {ind}")
    print(f"PC1: {x[ind]}")
    print(f"PC2: {y[ind]}")

# Create scatter plot with picking enabled
fig, ax = plt.subplots()
sc = ax.scatter(x, y, alpha=0.7, picker=True)  # Enable picking

ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_title("Interactive PCA Scatter Plot")

# Connect picking event
fig.canvas.mpl_connect('pick_event', onpick)

plt.show()
```

```python
import mplcursors

fig, ax = plt.subplots()
sc = ax.scatter(x, y, alpha=0.7)

ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_title("PCA Scatter Plot with Hover Selection")

# Enable hover & click selection
cursor = mplcursors.cursor(sc, hover=True)
cursor.connect("add", lambda sel: sel.annotation.set_text(f"PC1: {x[sel.index]:.2f}\nPC2: {y[sel.index]:.2f}"))

plt.show()
```
